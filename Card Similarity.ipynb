```python
# declare imports
import pandas as pd
import json
```

```python
# retrieve data
filename = "default-cards-20260323090744.json"
with open(filename, 'r', encoding='utf-8') as f:
    data = json.load(f)
df = pd.DataFrame(data)
```

```python
df['name_freq'] = df.groupby('name').id.transform('count')
df.drop_duplicates('name')[df.name_freq == 5].name.sample(5)
```
